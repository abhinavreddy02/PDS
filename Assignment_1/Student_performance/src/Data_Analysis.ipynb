```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Student_Data.csv")
```

```python
# Visualization 1: Histogram of Math Scores by Gender
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='math_score', hue='gender', kde=True)
plt.title('Histogram of Math Scores by Gender')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization 2: Boxplot of Reading Scores by Race/Ethnicity
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='race/ethnicity', y='reading_score')
plt.title('Boxplot of Reading Scores by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Reading Score')
plt.show()
```

```python
# Visualization 3: Bar Chart of Parental Level of Education
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='parental_level_of_education')
plt.title('Bar Chart of Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Visualization 4: Scatter Plot of Writing Score vs. Math Score
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='math_score', y='writing_score')
plt.title('Scatter Plot of Writing Score vs. Math Score')
plt.xlabel('Math Score')
plt.ylabel('Writing Score')
plt.show()
```

```python
# Visualization 5: Stacked Bar Chart of Lunch and Test Preparation Course
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='lunch', hue='test_preparation_course')
plt.title('Stacked Bar Chart of Lunch and Test Preparation Course')
plt.xlabel('Lunch')
plt.ylabel('Count')
plt.legend(title='Test Preparation Course', loc='upper right')
plt.show()
```
