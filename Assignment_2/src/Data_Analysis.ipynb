```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./clean_data.csv")
df
```

```python
from matplotlib import pyplot as plt
df['Year'].plot(kind='hist', bins=20, title='Year')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Bar chart of transmission types
transmission_counts = df['Transmission'].value_counts()
plt.figure(figsize=(8, 6))
transmission_counts.plot(kind='bar', color='green')
plt.xlabel('Transmission')
plt.ylabel('Number of Cars')
plt.title('Bar Chart of Transmission Types')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
# Pie chart of fuel types
fuel_type_counts = df['Fuel_Type'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(fuel_type_counts, labels=fuel_type_counts.index, autopct='%1.1f%%', startangle=140, colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'])
plt.axis('equal')
plt.title('Pie Chart of Fuel Types')
plt.show()
```

```python
# @title Question C
df = pd.get_dummies(df, columns=['Fuel_Type', 'Transmission'])
```

```python
df
```

```python
# @title Question D
from datetime import datetime
# Calculate current year
current_year = datetime.now().year

# Calculate the age of the car
df['Car_Age'] = current_year - df['Year']
```

```python
df.head()
```

```python
# Define function to categorize engine capacity
def categorize_engine_capacity(engine):
    if engine <= 1390:
        return 'Small'
    elif 1390 < engine <= 1950:
        return 'Medium'
    else:
        return 'Big'

# Apply the function to create the new column 'Engine_Capacity'
df['Engine_Capacity'] = df['Engine'].apply(categorize_engine_capacity)
```

```python
df.head()
```

```python
# Count the occurrences of each category
engine_capacity_counts = df['Engine_Capacity'].value_counts()

# Plot a bar chart
plt.figure(figsize=(8, 6))
engine_capacity_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Engine Capacity')
plt.xlabel('Engine Capacity')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

```python
# Plot a pie chart
plt.figure(figsize=(8, 6))
engine_capacity_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightgreen', 'lightblue', 'lightcoral'], startangle=140)
plt.title('Distribution of Engine Capacity')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
# @title Question E
# Select: Choose specific columns
selected_columns = df[['Name', 'Location', 'Year', 'Price']]

print("Selected Columns:")
print(selected_columns)
```

```python
# Filter: Select rows based on condition
filtered_data = df[df['Fuel_Type_Diesel'] == 1]

num_rows_filtered_data = len(filtered_data)


print("\nFiltered Data (Diesel cars only):")
print("Number of rows in the filtered data:", num_rows_filtered_data)
print(filtered_data)
```

```python
filtered_data = df[df['Engine_Capacity'] == 'Medium']

num_rows_filtered_data = len(filtered_data)

print("\nFiltered Data :")
print("Number of rows in the filtered data:", num_rows_filtered_data)
print(filtered_data)
```

```python
# Rename: Rename columns
renamed_df = df.rename(columns={'Unnamed: 0': 'ID', 'Kilometers_Driven': 'Km_Driven'})

print("\nRenamed Columns:")
print(renamed_df)
```

```python
# Mutate: Add or modify columns
df['Price_in_lakhs'] = df['Price'] * 100000

print("\nMutated DataFrame with Price in Lakh:")
print(df)
```

```python
# Scatter plot of Engine vs. Price
plt.figure(figsize=(10, 6))
plt.scatter(df['Engine'], df['Price'], color='blue', alpha=0.5)
plt.xlabel('Engine (CC)')
plt.ylabel('Price (in Lakhs)')
plt.title('Scatter Plot of Engine vs. Price')
plt.grid(True)
plt.show()
```

```python
# Arrange: Sort rows by one or more columns
sorted_yr = df.sort_values(by='Year', ascending=False)

print("\nSorted DataFrame by Year:")
print(sorted_yr)
```

```python
# Arrange: Sort rows by one or more columns
sorted_Mileage = df.sort_values(by='Mileage', ascending=False)

print("\nSorted DataFrame by Mileage of the car:")
print(sorted_Mileage)
```

```python
# Group by operations
grouped_summary = df.groupby('Location').agg({'Price': 'mean', 'Year': 'max'}).reset_index()
print("\nGrouped Summary by Location:")
print(grouped_summary)
```

```python
grouped_data = df.groupby(['Location', 'Year']).size().reset_index(name='Count')

print(grouped_data)
```

```python
import matplotlib.pyplot as plt
grouped_cars_by_location = df.groupby('Location').size().reset_index(name='Number_of_Cars')

print(grouped_cars_by_location)
```

```python
# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(grouped_cars_by_location['Number_of_Cars'], labels=grouped_cars_by_location['Location'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Number of Cars by Location')
plt.show()
```

```python
# Group by car age
grouped_data = df.groupby('Car_Age').size().reset_index(name='Cars_Count')

print("Grouped data by car age:")
print(grouped_data)
```

```python
# Plotting using Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(grouped_data['Car_Age'], grouped_data['Cars_Count'], color='skyblue')
plt.xlabel('Car Age')
plt.ylabel('Number of Cars')
plt.title('Number of Cars by Car Age')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
